```python
import numpy as np
import pandas as pd
```

```python
import sklearn
from sklearn.linear_model import LogisticRegression
```

```python
Dataframe= pd.read_csv("C:\\Users\\ACER\\OneDrive\\Desktop\\16 June_hackathon\\training_set_features.csv")
```

```python
Dataframe.columns
```

```python
xyz_variables= Dataframe.drop(columns=['respondent_id','doctor_recc_seasonal','opinion_seas_risk','opinion_seas_sick_from_vacc','opinion_seas_vacc_effective', 'employment_industry', 'employment_occupation'])
```

```python
output= pd.read_csv("C:\\Users\\ACER\\OneDrive\\Desktop\\16 June_hackathon\\training_set_labels.csv")
```

```python
output.columns
```

```python
y_xyz_train=output.drop(columns=['respondent_id','seasonal_vaccine'])
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
Cols=['age_group','education','race','sex','income_poverty','marital_status','rent_or_own','employment_status','hhs_geo_region','census_msa']
```

```python
xyz_variables[Cols]=xyz_variables[Cols].apply(le.fit_transform)
```

```python
xyz_variables.head(10)
```

```python
xyz_variables=xyz_variables.fillna(0)
Model=LogisticRegression()
Model=Model.fit(xyz_variables,y_xyz_train.values.ravel())
```

```python
Test= pd.read_csv("C:\\Users\\ACER\\OneDrive\\Desktop\\16 June_hackathon\\test_set_features.csv")
Test= Test.drop(columns=['respondent_id','doctor_recc_seasonal','opinion_seas_risk','opinion_seas_sick_from_vacc','opinion_seas_vacc_effective', 'employment_industry', 'employment_occupation'])
Cols=['age_group','education','race','sex','income_poverty','marital_status','rent_or_own','employment_status','hhs_geo_region','census_msa']
Test[Cols]=Test[Cols].apply(le.fit_transform)
Test=Test.fillna(0)
```

```python
y_predict=Model.predict_proba(Test)
y_predict
```

```python
y_predict_df=pd.DataFrame(y_predict,columns=[f'h1n1_vaccine_{i}' for i in Model.classes_])
y_predict_df.to_excel('xyz_vaccine predictions.xlsx',index=False)
```
